```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Importing the Daataset with Pandas
df = pd.read_csv('Data/Final_Weather.csv')
df
```

```python
# CHeecking the NULL Values
sns.heatmap(df.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

```python
# Replacing the Special Value ('NONE') with NaN values
df = df.replace('None',np.NaN)
df
```

```python
# CHeecking the NULL Values
sns.heatmap(df.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

```python
#Using Interpolation to cope up with NaN values.
new_df = df.dropna()
new_df
```

```python
sns.heatmap(new_df.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

```python
df = new_df
df
```

```python
df.to_csv('Data/dataset.csv', index=False)
```
